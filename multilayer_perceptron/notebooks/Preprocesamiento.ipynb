```python
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler


# Columnas del dataset
columnas = [
    "X_Minimum", "X_Maximum", "Y_Minimum", "Y_Maximum", "Pixels_Areas",
    "X_Perimeter", "Y_Perimeter", "Sum_of_Luminosity", "Minimum_of_Luminosity",
    "Maximum_of_Luminosity", "Length_of_Conveyer", "TypeOfSteel_A300",
    "TypeOfSteel_A400", "Steel_Plate_Thickness", "Edges_Index", "Empty_Index",
    "Square_Index", "Outside_X_Index", "Edges_X_Index", "Edges_Y_Index",
    "Outside_Global_Index", "LogOfAreas", "Log_X_Index", "Log_Y_Index",
    "Orientation_Index", "Luminosity_Index", "SigmoidOfAreas",
    "Pastry", "Z_Scratch", "K_Scatch", "Stains", "Dirtiness", "Bumps", "Other_Faults"
]
```

```python
#PATHS
def find_project_root(start_path: Path):
    """
    Busca la raíz del proyecto recorriendo directorios hacia arriba.
    """
    for parent in [start_path] + list(start_path.parents):
        if (parent / "data").exists():
            return parent

    raise FileNotFoundError("No se encontró la carpeta 'data'.")


def construir_rutas():
    """
    Construye todas las rutas necesarias del proyecto.
    """
    current_dir = Path.cwd()
    root_dir = find_project_root(current_dir)

    raw_path = root_dir / "data" / "raw" / "Faults.NNA"
    processed_dir = root_dir / "data" / "processed"
    processed_file = processed_dir / "faults_processed.csv"

    processed_dir.mkdir(parents=True, exist_ok=True)

    return current_dir, root_dir, raw_path, processed_dir, processed_file
```

```python
# Carga de datos
def cargar_datos(raw_path: Path):
    """
    Carga el dataset desde archivo.
    """
    if not raw_path.exists():
        print(f"Error: no se encontró el archivo en:\n{raw_path}")
        return None

    df = pd.read_csv(raw_path, sep=r"\s+", names=columnas)
    return df
```

```python
# Preprocesamiento
def separar_variables(df):
    """
    Separa características (X) y etiquetas (y).
    """
    X = df.iloc[:, :27]  # variables de entrada
    y = df.iloc[:, 27:]  # etiquetas
    return X, y


def normalizar_datos(X):
    """
    Normaliza las variables de entrada.
    """
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    X_scaled_df = pd.DataFrame(X_scaled, columns=columnas[:27])
    return X_scaled_df


def unir_datos(X_scaled_df, y):
    """
    Une datos normalizados con etiquetas.
    """
    df_final = pd.concat([X_scaled_df, y.reset_index(drop=True)], axis=1)
    return df_final
```

```python
# Guardar datos
def guardar_datos(df_final, processed_file):
    """
    Guarda el dataset procesado en CSV.
    """
    df_final.to_csv(processed_file, index=False)
    print("Guardado en:", processed_file)
```

```python
# Ejecución
def normalizar_y_limpiar():
    """
    Ejecuta todo el pipeline de procesamiento.
    """
    current_dir, root_dir, raw_path, processed_dir, processed_file = construir_rutas()

    print("Directorio actual:", current_dir)
    print("ROOT detectado:", root_dir)

    df = cargar_datos(raw_path)

    if df is None:
        return

    X, y = separar_variables(df)
    X_scaled_df = normalizar_datos(X)
    df_final = unir_datos(X_scaled_df, y)

    guardar_datos(df_final, processed_file)


# Ejecutar
normalizar_y_limpiar()
```
